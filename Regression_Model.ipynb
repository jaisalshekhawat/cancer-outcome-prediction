```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Drop irrelevant identifier columns
df2 = df.drop(columns=["Patient_ID", "Death_Date", "Cause_of_Death", "Primary_Physician"])

# Fill missing values
df2 = df2.fillna("Unknown")

# Convert date columns to datetime
df2["Diagnosis_Date"] = pd.to_datetime(df2["Diagnosis_Date"])
df2["Treatment_Start_Date"] = pd.to_datetime(df2["Treatment_Start_Date"])

# Feature: days between diagnosis and treatment
df2["Days_To_Treatment"] = (df2["Treatment_Start_Date"] - df2["Diagnosis_Date"]).dt.days

# Drop original date columns
df2 = df2.drop(columns=["Diagnosis_Date", "Treatment_Start_Date"])

# Separate target and features
y = df2["Outcome"]
X = df2.drop(columns=["Outcome"])

# One-hot encode categorical features
X = pd.get_dummies(X)

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

<h2>Train a Model</h2>

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Report:\n", classification_report(y_test, y_pred))
```

<h2>Model’s Performance</h2>
<ul>
    <li>Overall Accuracy: 0.54 → This means your model correctly predicted 54% of the outcomes.</li>
</ul>

<h4>Class-wise Performance:</h4>
<ul>
    <li>Deceased: Almost perfect. precision=0.98, recall=1.00 — the model almost always gets this right.</li>
    <li>Recovered and Under Treatment: Much lower performance. Precision and recall are in the mid-40s to 50s.</li>

</ul>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Handle missing values if any
import numpy as np
X = np.nan_to_num(X)

# Split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)

# Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Predict & Evaluate
y_pred = knn.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Report:\n", classification_report(y_test, y_pred))
```
